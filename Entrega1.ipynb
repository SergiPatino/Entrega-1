```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcio(u):
   f = np.sin(u)
   g = np.cos(u)
   h = np.tan(u)
   return f,g,h
```

```python
x = np.arange(0, 2*np.pi, 0.1)
```

```python
f,g,h = funcio(x)
```

```python
plt.plot(x, f, 'r', x, g, 'b', x, h, 'g')
plt.plot(x, f, 'ro', x, g, 'bo', x, h, 'go')
plt.xlabel('Valors de x')
plt.ylabel('Valors de y')
plt.title('Representaci√≥ de y=cos(x), y=sin(x) i y=tan(x)')
plt.legend(['sin(x)', 'cos(x)','tan(x)'])
plt.grid(True)
plt.ylim((-5,5))
plt.show()
```
